```python
import pandas as pd
```

```python
df=pd.read_csv('mobile_train.csv')
```

```python
df
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import show
```

```python
for i in df.columns:
  if df[i].dtypes!=object:
    sns.scatterplot(data=df,x=i,y='price_range')
    print(i,'vs price_range')
    show()
```

```python
#  sepreat input and output features
x=df.drop(labels=['price_range'],axis=1)
y=df[['price_range']]
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
# train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
#
from keras.utils import to_categorical
```

```python
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
ytrain_cat[:5]
```

```python
ytest_cat[:5]
```

```python
# Building networks
from keras.models import Sequential
from keras.layers import Input,Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
# initialize the model
model=Sequential()

# input layer
model.add(Input(shape=(xtrain.shape[1])))

# create hidden layer
model.add(Dense(units=64,activation='relu'))

# add dropout layer
model.add(Dropout(rate=0.3))

# create a hidden layer 2
model.add(Dense(units=32,activation='relu'))

# add dropout layer
model.add(Dropout(rate=0.2))

# output layer
model.add(Dense(units=ytrain_cat.shape[1],activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
es=EarlyStopping(monitor='val_loss',patience=10)
```

```python
hist=model.fit(xtrain,ytrain_cat,validation_split=0.2,epochs=200,callbacks=[es])
```

```python
# curve
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'],label='Trainingloss')
plt.plot(hist.history['val_loss'],label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('categorical crossentropy')
plt.title('Learning Curve')
plt.show()
```

```python
# evalute model
model.evaluate(xtrain,ytrain_cat)
```

```python
# evalute model
model.evaluate(xtest,ytest_cat)
```

```python
# predict the results fro train and test
# it will pridict the probability
yprob_train = model.predict(xtrain)
yprob_test = model.predict(xtest)
```
