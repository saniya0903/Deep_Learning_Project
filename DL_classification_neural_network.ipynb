```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df.info()
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# separate input and output feature
x=df.drop(labels=['id','booking_status'],axis=1)
y=df[['booking_status']]
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x=pd.DataFrame(ss.fit_transform(x),columns=x.columns)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
# Build nerul network
from keras.models import Sequential
from keras.layers import Input,Dense,Dropout
```

```python
# initialize the model
model=Sequential()

# input layer
model.add(Input(shape=(xtrain.shape[1])))

# create hidden layer
model.add(Dense(units=64,activation='relu'))

# add dropout layer
model.add(Dropout(rate=0.5))

# create a hidden layer 2
model.add(Dense(units=32,activation='relu'))

# add dropout layer
model.add(Dropout(rate=0.4))

# output layer
model.add(Dense(units=1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])        # binary cross entropy or logg loss function
```

```python
hist=model.fit(xtrain,ytrain,validation_split=0.2,epochs=60)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(hist.history['loss'],label='Train Loss')
plt.plot(hist.history['val_loss'],label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Binary Cross Entropy')
plt.title('Learning Curve')
plt.legend()
plt.show()
```

```python
# evalute model
model.evaluate(xtrain,ytrain)
```

```python
tr_prob=model.predict(xtrain)
ts_prob=model.predict(xtest)
```

```python
# sigmoid will give probabilities
tr_prob[:5]
```

```python
ts_prob
```

```python
tr_pred=[]
for i in tr_prob:
  if i>=0.5:
    tr_pred.append(1)
  else:
    tr_pred.append(0)
```

```python
tr_pred[:5]
```

```python
ts_pred=[]
for i in ts_prob:
  if i>=0.5:
    ts_pred.append(1)
  else:
    ts_pred.append(0)
```

```python
ts_pred[:5]
```

```python
ytrain.head()
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

```python
cf=confusion_matrix(ytest,ts_pred)
cfd=ConfusionMatrixDisplay(cf)
cfd.plot()
```

```python
# saving model
model.save('BookingModel.keras')
```
